## Wrist shot model and derived statistic

```python
%matplotlib inline
import os
import sys
import numpy 
import pandas 
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

# !pip install plotly

# imports regression library
# OLS: ordinary least squares (estimation technique used to estimate the linear regression model)
import statsmodels.api as sm
from statsmodels.formula.api import ols
    
# Set some Pandas options
pandas.set_option('display.notebook_repr_html', True)
pandas.set_option('display.max_columns', 20)
pandas.set_option('display.max_rows', 50)

from decimal import getcontext, Decimal
# Set the precision.
getcontext().prec = 2
```

```python
dm = pandas.read_csv('shots.csv')
dm = dm[(dm['advantagetypeshooter']=='EV')] 
dm = dm[dm['zone']=="O"]
```

Create variables

```python
dm['isgoal'] = dm['eventtype']=='GOAL'
dm['XNormAdj'] = 99 - dm['XNorm']
dm.groupby(['eventtype'])[['XNormAdj', 'YNorm']].describe()
```

```python
dm['XNormAdj'] = dm['XNormAdj'] + 1
dm['YNormAdj'] = numpy.absolute(dm['YNorm'])
dm['YNormAdj'] = dm['YNormAdj'] + 1

dm['lnX'] = numpy.log(dm['XNormAdj'])
dm['lnY'] = numpy.log(dm['YNormAdj'])

dm['dscore'] = dm['tgoals'] - dm['ogoals']
dm['oteamcode'] = dm.apply(lambda x: 1 if (x['teamcode'] == x['hteamcode']) else x['vteamcode'], axis=1)
```

Create wrist shot specific data frame

```python
dw = dm[dm['shotType']=='Wrist']
```

```python
dg = dw[dw['eventtype']=='GOAL']

plt.hist2d(dg['XNormAdj'], dg['YNorm'],bins=30)
plt.show()
```

Logit model estimation and predictions

    * m1: length
    * m2: width
    * m3: length and width

Linear probability models (i.e., OLS) can predict probabilties outside of the 0-1 interval. A logistic regression estimates the log odds relationship. The resultant probabilities are bound between 0 and 1: 

$$ p_i = \frac{1}{1+exp(-X'\beta)}$$

The paramater estimates are in terms of log odds. Marginal effects represent unit change increments similar to OLS.

Model the probability that a shot results in a goal in terms of length.

```python
m1 = sm.Logit(dw['isgoal'], sm.add_constant(dw['lnX'])).fit()
print m1.get_margeff(method='dydx', at='overall').summary()

# Generate predicted values and predict results
dw.ix[:, 'pgoalx'] = m1.predict()
dw.plot(kind='scatter',x='XNormAdj', y='pgoalx')
```

Each percent increase in length from the net decrease the probability of a goal by 0.09.

Model the probability that a shot results in a goal in terms of width

```python
m2 = sm.Logit(dw['isgoal'], sm.add_constant(dw['lnY'])).fit()
print m2.get_margeff(method='dydx', at='overall').summary()

# Generate predicted values and predict results
dw.ix[:, 'pgoaly'] = m2.predict()
dw.plot(kind='scatter', x='YNormAdj', y='pgoaly')
```

Each percent increase in width decreases the probability of a goal by 0.05.

Model the probability that a shot results in a goal in terms of width

```python
m3 = sm.Logit(dw['isgoal'], sm.add_constant(dw[['lnX', 'lnY']])).fit()
print m3.get_margeff(method='dydx', at='overall').summary()
dw.ix[:, 'pgoal'] = m3.predict()
```

Each percent increase in length and width decreases the probability of a goal by 0.80 and 0.05, respectively.
