Create Variables 

```python
dm['isgoal'] = dm['eventtype']=='GOAL'
dm['XNormAdj'] = 99 - dm['XNorm']
dm.groupby(['eventtype'])[['XNormAdj', 'YNorm']].describe()
```

```python
dm['XNormAdj'] = dm['XNormAdj'] + 1 
dm['YNormAdj'] = numpy.absolute(dm['YNorm'])
dm['YNormAdj'] = dm['YNormAdj'] + 1 

dm['lnX'] = numpy.log(dm['XNormAdj'])
dm['lnY'] = numpy.log(dm['YNormAdj'])

dm['dscore'] = dm['tgoals'] - dm['ogoals']
dm['oteamcode'] = dm.apply(lambda x: 1 if (x['teamcode'] == x['hteamcode']) else x['vteamcode'], axis=1) 
```

Create wrist shot specific data frame 

```python
ds = dm[dm['shotType']=='Slap']
db = dm[dm['shotType']=='Backhand']
```

```python
dg1 = ds[ds['eventtype']=='GOAL']

plt.hist2d(dg1['XNormAdj'], dg1['YNorm'],bins=30)
plt.show
```

Logit model estimation and predictions
- ia) m1: length (SLAP)
- ib) m11: length (BACKHAND)
- iia) m2: width (SLAP)
- iib) m22: width (BACKHAND)
- iiia) m3: length and width (SLAP)
- iiib) m33: length and width (BACKHAND)


ia) Model the probability that a SLAP shot results in a goal in terms of length

```python
m1 = sm.Logit(ds['isgoal'], sm.add_constant(ds['lnX'])).fit()
print m1.get_margeff(method='dydx', at='overall').summary()

#Generate predicted values and predicted results
ds.ix[:, 'pgoalx'] = m1.predict()
ds.plot(kind='scatter',x='XNormAdj', y='pgoalx')
```

Observation/Result: Each percent increase in length from the net decrease the probability of a goal by 0.06
